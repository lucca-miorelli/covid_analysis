No ano de 2020, o mundo inteiro foi forçado a mudar hábitos. Alguns muito comuns antes tiveram de ser modificados devido à crise na saúde mundial, em decorrência da pandemia de COVID-19.

A COVID-19 é uma doença infecciosa causada pelo coronavírus da síndrome respiratória aguda grave 2 (SARS-CoV-2).

Transmitido principalmente por meio de gotículas provenientes de tosses ou espirros de pessoas infectadas, a gravidade dos sintomas varia muito de pessoa para pessoa.

<center><img alt="covid-analysis-banner" width="65%" src="https://github.com/lucca-miorelli/covid_analysis/raw/main/covid-analysis-banner.jpg"></center>

O que foi observado é que cerca de 80% dos casos confirmados são assintomáticos e rápidos, os quais se recuperam sem nenhuma sequela.

No entanto, aproximadamente 15% podem desenvolver infecções graves e precisarão de oxigênio. O restante dos infectados, que representam 5%, serão classificadas como infecções muito graves e precisarão de ventilação assistida, por meio de respiradores mecânicos em ambiente hospitalar. 

No ápice da pandemia, a grande questão é justamente a disponibilidade de assistência médica, pois não é possível atender simultaneamente grande quantidade de pessoas infectadas. A assistência médica, que é essencial para casos mais graves da doença, acabava se tornando inviável e não sobrava outra alternativa.

Passados alguns meses de pandemia, felizmente a ciência agiu rapidamente, tendo decifrado um dos maiores desafios do último século: desenvolver a vacina contra o vírus, contribuindo para a diminuição de mortes de COVID-19 e - no médio e longo prazo - velhos hábitos possam voltar.

Com o objetivo de elevar a consciência situacional a respeito do COVID-19 no Brasil, esta é uma análise sobre os dados públicos da evolução da doença no Brasil e no mundo.

## Obtenção dos Dados

Para esta análise, foram utilizados os dados disponíveis no repositório aberto  [*Our World in Data*](https://ourworldindata.org/coronavirus), que é uma iniciativa cujo objetivo principal é disponibilizar dados de problemas mundiais à comunidade de forma acessível.

> **"Research and data to make progress against the world’s largest problems"**

Os dados relativos à COVID-19 são coletados com periodicidade e em fontes oficiais das nações mundiais, desde o início de suas divulgações. Assim, podemos obter análises de:
* Casos confirmados;
* Mortes;
* Vacinação.

Além de possuir outros dados - como `PIB per capita`, `unidades de tratamento hospitalar por mil habitantes`, `população`, `índice de desenvolvimento humano`, entre outros - o que torna mais fácil analisar possíveis correlações.

Apesar disso, é importante ressaltar a dificuldade de analisar uma doença a qual a própria ciência ainda não a desvendou por completo, e com divergências de dados entre governos federal, estadual e fontes não oficiais por exemplo, além de alguns países não fornecerem dados por motivos políticos, de infraestrutura ou até mesmo por ataques *hackers*.


### Dicionário de Variáveis
O dataset possui mais de 65 variáveis, no entando ressalto as que mais utilizei na análise:

#### Variáveis de casos confirmados
| Variável                       | Descrição                                                         |
|:---------------------------------|:----------------------------------------------------------------------|
| `total_cases`                    | Total de casos confirmados de COVID-19                                     |
| `new_cases`                      | Novos casos confirmados de COVID-19                                       |
| `new_cases_smoothed`             | Média móvel de 7 dias de casos confirmados de COVID-19                      |
| `total_cases_per_million`        | Total de casos confirmados de COVID-19 por 1.000.000 de habitantes                |

### Variáveis de mortes
| Variável                          | Descrição                                                             |
|:----------------------------------|:------------------------------------------------------------------------|
| `total_deaths`                    | Total de mortes de COVID-19                                     |
| `new_deaths`                      | Novas mortes de COVID-19                                       |
| `new_deaths_smoothed`             | Média móvel de 7 dias de mortes de COVID-19                      |
| `total_deaths_per_million`        | Total de mortes de COVID-19 por 1.000.000 de habitantes               |
| `new_deaths_per_million`          | Novas mortes de COVID-19 por 1.000.000 de habitantes                  |

### Vacinação
| Varriável                                    | Descrição                                                  
|:---------------------------------------------|:--------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| `people_vaccinated`                          | Total de pessoas que receberam no mínimo 1 dose da vacina dose                                                                                                                                                                                                                                                                                     |
| `people_fully_vaccinated`                    | Total de pessoas com o esquema de vacinação completo                                                                         
| `new_vaccinations_smoothed_per_million`      | Média móvel de 7 dias do total de vacinas aplicadas por 1.000.000 de habitantes                                                                                                                                                                                                                                         |
 
### Others
| Variável                     | Descrição                                                                                                                                                                                                                                |
|:-----------------------------|:-------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------------|
| `iso_code`                   | ISO 3166-1 alpha-3 – código de três letras referente ao país                                                                                                                                                                                            |
| `continent`                  | Continent ao qual pertence o local geográfico                                                                                                                                                                                                      |
| `location`                   | Nome do local geográfico                                                                                                                                                                                                                       |
| `date`                       | Data de observação                                                                                                                                                                                                                        |
| `population`                 | População (valores mais atualizados). Ver [repositório](https://github.com/owid/covid-19-data/blob/master/scripts/input/un/population_latest.csv) para lista completa de fontes |
| `aged_70_older`              | Parcela da população com idade igual ou maior que 70 anos em 2015                                                                                                                                                                                 |
| `gdp_per_capita`             | Produto interno bruto em dólares, dados de 2011                                                                                                                    |
| `hospital_beds_per_thousand` | Leitos de hospital por 1.000 de habitantes, últimos dados de 2010                                                                                                                                                                      |
| `life_expectancy`            | Expectativa de vida ao nascimento em 2019                                                                                                                                                                                                           |
| `human_development_index`    | Índice de desenvolvimento humano do país, atualizados em 2019. Fonte: http://hdr.undp.org/en/indicators/137506 |

Além destas variáveis, todo o dataset é bem documentado no [repositório oficial](https://github.com/owid/covid-19-data/tree/master/public/data/owid-covid-codebook.csv), que é disponível ao público.

**Evolução de casos confirmados no mundo**

Abaixo, os gráficos expõem a evolução da pandemia no mundo ( com suas ondas características de contaminação) e dados sobre as variantes até agora elencadas pela *World Health Organization*.

* As datas de cada variante consideradas nesta análise são as datas nas quais a WHO passou a considerar a variante como **VOC (Variant of Concern)** - categoria de **variante de preocupação** - e não mais como *VOI (Variant of Interest)*, categoria de variante que existe, mas ainda não apresenta ameaça.

Mais informações das variantes do vírus no site da [WHO](https://www.who.int/en/activities/tracking-SARS-CoV-2-variants/).

---

**IMPORTANTE:**  
As cores abaixo servem para ilustrar a data de início da variante como variante de preocupação, mas a circulação de variantes não é exclusiva, ou seja, as variantes circulam simultaneamente. A relação da data com ondas de contaminação é clara, mas não significa ser a causa.

```
variant_dates = {
    'alpha_beta':'2020-12-18',
    'gamma':'2021-01-11',
    'delta': '2021-05-11',
    'omicron': '2021-11-26',
    }

# dataframe com datas no index
df2 = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv', index_col="date", parse_dates=True)
df_world = df2.loc[df2.location == 'World', ['new_deaths_smoothed', 'new_cases_smoothed']]

# create plot moving average de casos no mundo
fig, ax = plt.subplots(figsize=(21,9), dpi=72)

df_world.loc['2020-02-20':'2022-01-09', 'new_cases_smoothed'].plot(kind='line', ax=ax, lw=5, label='Média móvel 7 dias', color = BLACK)

# customize spines
ax.spines['right'].set_visible(False)
ax.spines['top'].set_visible(False)

# customize labels and axis
ax.set_ylabel('')
ax.set_xlabel('')
ax.minorticks_off()
ax.set_ylim(0, top=None)
plt.yticks([i * 500000 for i in range(6)], ['0', '500.000', '1.000.000', '1.500.000', '2.000.000', '2.500.000'], fontsize=14)
ax.set_xticks(['2020-04-01', '2020-07-01', '2020-10-01', '2021-01-01', '2021-04-01', '2021-07-01', '2021-10-01', '2022-01-01'])
ax.set_xticklabels(['Apr', 'Jul', 'Oct', 'Jan\n2021', 'Apr', 'Jul', 'Oct', 'Jan\n2022'], fontsize=15)
ax.set_xlabel('')

# Fill between
ax.fill_between(df_world.index, df_world['new_cases_smoothed'], where=(df_world.index >= variant_dates['omicron']), color = 'grey', alpha=0.6) 
ax.fill_between(df_world.index, df_world['new_cases_smoothed'], where=(df_world.index <= variant_dates['omicron']), color='grey', alpha=0.4) 
ax.fill_between(df_world.index, df_world['new_cases_smoothed'], where=(df_world.index <= variant_dates['delta']), color='grey',alpha=0.3) 
ax.fill_between(df_world.index, df_world['new_cases_smoothed'], where=(df_world.index <= variant_dates['gamma']), color='grey', alpha=0.2) 

ax.text(x=variant_dates['omicron'], y=100000 , s=' Omicron', fontsize=16, color='white')
ax.text(x=variant_dates['delta'], y=100000,  s=' Delta', fontsize=20, color='white')
ax.text(x=variant_dates['gamma'], y=100000 , s=' Gama', fontsize=20, color='white')
ax.text(x='2020-09-10', y=100000, s='Alpha/Beta', fontsize=20, color='white')




# title and subtitle
ax.text(0, 1.15,'Evolução de casos confirmados no mundo',transform=ax.transAxes, fontsize=36, fontweight='bold',color='#58555A')
ax.text(0, 1.09,'Média móvel de casos confirmados em 7 dias.', transform=ax.transAxes, fontsize=21, color='#A1A5A6')

# Add caption
fig.text(
    0.12, 0.03, "Fonte: Our World In Data", color=GREY,
    fontsize=17, fontfamily="Econ Sans Cnd"
)

# Add caption
fig.text(
    0.12, 0.0, "https://ourworldindata.org/coronavirus", color=GREY,
    fontsize=18, fontfamily="Econ Sans Cnd"
)
```


No gráfico, podemos observar:
* Existem **ondas quase periódicas de contaminação** que coincidem com a descoberta de novas variantes.
* **No final de 2021 e início de 2022, houve grande registro de novos casos no mundo inteiro**. Além da maior circulação de pessoas em comparação ao ano anterior, isso também pode ser reflexo de uma cultura de testagem mais bem assentada que anos meses anteriores.

## Análise de correlação entre mortalidade e outros fatores

A seguir, foi realizada uma análise de correlação linear entre o número de mortes e alguns fatores como:
* Expectativa de vida do país;
* Números de leitos de hospital por mil habitantes;
* Índice de Desenvolvimento Humano (IDH);
* PIB per capita ;
* Parcela da população com 70 anos ou mais.

```
# análise de correlação de variáveis com mortalidade

#Data preparation

variables = ['location', 
             'total_deaths', 
             'total_cases', 
             'hospital_beds_per_thousand',
             'life_expectancy', 
             'human_development_index',
             'gdp_per_capita',
             'aged_70_older']

i = 0

for cont in continents:
  
  df_cont = pd.read_csv('https://covid.ourworldindata.org/data/owid-covid-data.csv', index_col="date", parse_dates=True)
  df_cont = df_cont.loc[df_cont.continent==cont]

  df_int = df_cont.loc['2021-12-31', variables]
  df_int['death_pctg'] = df_int['total_deaths'] / df_int['total_cases']   # percentage of death when infected
  df_int.drop('location', axis=1, inplace=True)

  if i == 0:
    df_corr = df_int.copy()
  else:
    df_corr = df_corr.append(df_int)
  
  i+=1

df_corr.reset_index().drop('date', axis=1, inplace=True)
df_corr.dropna(axis=0, inplace=True)
```

Para obter resultados mais confiáveis, **os dados utilizados foram normalizados**.
Essa é a forma mais correta de realizar a análise, pois as diferentes variáveis numéricas - que possuem escalas diferentes entre si - são trazidas a um mesmo intervalo e assim passam a não interferir menos no resultado. Outro fator importante é que com esse método não se assume qualquer distribuição.

Aqui está um [link](https://www.analyticsvidhya.com/blog/2020/04/feature-scaling-machine-learning-normalization-standardization/) para saber mais sobre normalização e padronização, muito utilizados em análises matemáticas e principalmente Machine Learning.

```
# data normalization with sklearn
from sklearn.preprocessing import MinMaxScaler

# fit scaler on training data
norm = MinMaxScaler().fit(df_corr)

# transform training data
corr_norm = norm.transform(df_corr)
df_corr_norm = pd.DataFrame(corr_norm, columns=df_corr.columns)
```

```
import seaborn as sns
sns.heatmap(df_corr_norm.corr(), annot=True, fmt='.2g', cmap='coolwarm', vmin=-1, vmax=1)
```

Com esta matriz de correlação linear, é possível ver que:
* Nenhuma variável demonstrou coeficiente de correlação linear muito relevante com `total_deaths`. A mais relevante foi `human_development_index` com somente 0.16.
* Por outro lado, PIB per capita (`gdp_per_capita`) apresentou coeficiente de -0.3 com `death_pctg`, variável que calculei do percentual de letalidade da doença (número de mortes dividido pelo número de casos confirmados). 
  * Apesar de ser um coeficiente relativamente baixo para pensar em uma relação causa/efeito, isso demonstra que é possível que **países com PIB per capita menor tenham taxa de letalidade maior**.


# **Vacinação**

```
from datetime import datetime as dt

# data da primeira vacina no brasil
bra_first_vac = df_bra.loc[df_bra.people_vaccinated != 0].dropna(subset=['people_vaccinated'], axis=0)[:1].index.tolist()[0]
print("A primeira vacina no Brasil foi em {}\n".format(bra_first_vac.strftime("%d %b %Y")))

# porcentagem de vacinados no dia da máxima média móvel de mortes
df_bra['vaccinated_percentage'] = df_bra['people_fully_vaccinated'] / df_bra['population']

mov_death_max = df_bra.new_deaths_smoothed.max()
day_max = df_bra.loc[df_bra.new_deaths_smoothed==mov_death_max].index.tolist()[0]
pctg_day_max = df_bra.loc[day_max].vaccinated_percentage * 100

print_str = "Somente {0:.2f}% da população brasileira estava com o esquema vacinal completo quando o país atingia a média de mortes máxima, cujo valor foi de {1:.0f} mortes, no dia {2}.\n"
print(print_str.format(pctg_day_max, mov_death_max, day_max.strftime('%d %b %Y')))

# data dos 15%, 25% e 50% de completamente vacinados
for pctg in [0.15, 0.25, 0.50]:
  date_vac = df_bra[df_bra.vaccinated_percentage >= pctg][:1].index.tolist()[0]
  print(print("Em {}, {}% da população brasileira estava com o esquema vacinal completo.\n".format(date_vac.strftime('%d %b %Y'), pctg * 100)))

# porcentagem de vacinados ao final de 2021
print("{:.2f}% da população brasileira estava o com esquema vacinal completo ao final de 2021.\n".format(df_bra.loc['2021-12-31', 'vaccinated_percentage'] * 100))
```
